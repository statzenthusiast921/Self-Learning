```python
#Python visualizations - 12.26.19
import numpy as np
import pandas as pd

import os

os.getcwd()

#os.chdir('C:\\Users\\Jon Zimmerman\\Desktop\\Python')
```

```python
df2=pd.read_csv('Absenteeism_at_work.csv',sep=';')
```

```python
df2.head()
```

```python
#pandas built in visualization capabilities are built on top of matplot lib
#if you look at the output, you will see matplotlib
#also called box and whisker plot
#each line represents quartile (25%)
df2.boxplot(column="Reason for absence")
#Historgram
df2['Reason for absence'].hist()
#X-Y Plots
df2.plot(x="Distance from Residence to Work",y="Age",kind="line")
df2.plot(x="Distance from Residence to Work",y="Age",kind="scatter")
```

```python
#KDE=Kernel Density Estimation Plot
df2.plot(x="Weight",y="Body mass index",kind="kde")
#continuous curve is drawn at these data points and then added
#together to create this smooth line graph, used when 
#histograms not appropriate
```

```python
#MATPLOT LIB
#https://matplotlib.org/gallery/index.html

import matplotlib.pyplot as plt
#np.array or np.ma.masked_array as input
#pandas data objects and np.matrix may or may not work
#best to convert these to np.array objects prior to plotting

#you will need to add plt.show() if you are not using jupyter notebook
```

```python
#few important commands which can be used with any graphs
fig=plt.figure() #an empty figure with no axes
fig.suptitle('No axes on this figure') #add a title
fig,ax_list=plt.subplots(2,2) #a figure with a 2X2 grid of axes

x=df2['Weight']
y=df2['Body mass index']
plt.plot(x,y,label='linear')

#fig is the figure instance
fig,ax=plt.subplots()
ax.plot(x,y)

import seaborn as sns
sns.set(style="ticks")

sns.pairplot(df2)

#Run all this together
import matplotlib.pyplot as plt
sns.set(style="ticks")
```

```python
#Initialize the figure
f,ax=plt.subplots(figsize=(7,6))
ax.set_xscale("log")

#Plot the orbital period with horizontal boxes
sns.boxplot(x="Reason for absence", y="Absenteeism time in hours",data=df2)

#Add in points to show each observation
sns.swarmplot(x="Reason for absence",y="Absenteeism time in hours",data=df2,size=2,color=".3",linewidth=0)

#Tweak the visual presentation
ax.xaxis.grid(True)
ax.set(ylabel="")
sns.despine(trim=True,left=True)
```

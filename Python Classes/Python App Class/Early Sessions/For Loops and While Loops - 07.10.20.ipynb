```python
#Looping through a Dictionary
```

```python
student_grades = {"Mary": 9.1,"Jon":8.8,"Amelia":7.6}
for grades in student_grades.items():
    print(grades)
```

```python
student_grades = {"Mary": 9.1,"Jon":8.8,"Amelia":7.6}
for grades in student_grades.keys():
    print(grades)
```

```python
student_grades = {"Mary": 9.1,"Jon":8.8,"Amelia":7.6}
for grades in student_grades.values():
    print(grades)
```

```python
phone_numbers = {"John Smith": "+37682929928", "Marry Simpons": "+423998200919"}
for i, j in phone_numbers.items():
    print("%s: %s" % (i,j))
```

```python
phone_numbers = {"John Smith": "+37682929928", "Marry Simpons": "+423998200919"}
for i in phone_numbers.values():
    print(i)
```

```python
phone_numbers = {"John Smith": "+37682929928", "Marry Simpons": "+423998200919"}
for i in phone_numbers.values():
    print(i.replace("+","00"))
```

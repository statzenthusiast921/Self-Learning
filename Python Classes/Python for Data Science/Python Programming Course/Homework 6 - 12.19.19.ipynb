```python
from matplotlib import pyplot as plt
import seaborn as sns
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
data.columns=['DayofWeek','Director','Genre','MovieTitle','ReleaseDate',
              'Studio','AdjGrossMil','BudgetMil','GrossMil','IMDB_Rating',
              'MovieLens_Rating','OverseasMil','OverseasPerc','ProfitMil','Profit_Perc',
              'RuntimeMin','USMil','GrossPercMil']
```

```python
data.head()
```

```python
a=sns.boxplot(data=data,x='Genre',y='GrossPercMil')
```

```python
data.Genre.unique()
```

```python
data.Studio.unique()
```

```python
#Need to filter by genre and studio
#Genres to include: action, comedy, adventure, animation, drama
#Studios to include: Buena Vista, Sony, Universal, WB, Paramount, and Fox

genrefilters=(data.Genre=="adventure")|(data.Genre=="comedy")| \
             (data.Genre=="action")|(data.Genre=="animation")| \
             (data.Genre=="drama")

studiofilters=(data.Studio=="Buena Vista Studios")| \
              (data.Studio=="Sony")| \
              (data.Studio=="Universal")| \
              (data.Studio=="WB")| \
              (data.Studio=="Paramount Pictures")| \
              (data.Studio=="Fox")
```

```python
data2=data[genrefilters & studiofilters]
```

```python
len(data2)
```

```python
sns.set_style("darkgrid")
a2=sns.boxplot(data=data2,x='Genre',y='GrossPercMil',color='lightgray')
a2=sns.stripplot(x="Genre",y='GrossPercMil',data=data2,
                 hue="Studio")
plt.title("Domestic Gross % by Genre",fontsize=20)
plt.ylabel("Gross % US",fontsize=20)
plt.xlabel("Genre",fontsize=20)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0,
           facecolor='none',edgecolor='none')
plt.show()
```
